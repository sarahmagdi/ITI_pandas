#### Read XML dataset that has multilevel children

```python
df = pd.read_xml('../datasets/books_multilevel_children.xml')
df.head()
```

##### Not as expected, so let's try XPATH as it may help here

```python
df = pd.read_xml('../datasets/books_multilevel_children.xml', xpath='//book | //book/child')
df.head()
```

##### Not as expected at all :(, so let's try to do a workaround

```python
df1 = pd.read_xml('../datasets/books_multilevel_children.xml')
del df1['child']
df1.head()
```

```python
df2 = pd.read_xml('../datasets/books_multilevel_children.xml', xpath='//book/child')
df2.head()
```

```python
df = pd.concat([df1, df2], axis=1)
df.head()
```
