```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/surveys.csv')
df.head()
```

```python
df.info()
```

##### drop rows that have missing data

```python
df_2 = df.dropna(how='any')
df_2.head()
```

```python
df_2.info()
```

```python
df_3 = df.dropna(how='any', subset=['weight'])
df_3.head()
```

```python
df_3.info()
```

```python
df_3_1 = df.dropna(thresh=6)
df_3_1.info()
```

```python
df_3_2 = df.dropna(thresh=2, subset=['weight', 'hindfoot_length'])
df_3_2.info()
```

## fill-in missing data

```python
df.info()
```

```python
df_4 = df.fillna(value='0')
df_4
```

#### what happened seems not the best for categorical fields!

```python
df_4.info()
```

```python
filling_values = {"weight": 0, "hindfoot_length": df['hindfoot_length'].mean(), "sex": df['sex'].mode().iloc[0], "species_id": df['species_id'].mode().iloc[0]}
df_5 = df.fillna(value=filling_values)
df_5
```

#### Now, I'm more staified

```python
df_5.info()
```

```python
df_5.to_csv("../datasets/surveys_2.csv", index=None)
```
