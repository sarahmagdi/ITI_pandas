```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/surveys.csv')
df.info()
```

```python
df
```

```python
pd.get_dummies(df['sex'])
```

```python
pd.get_dummies(df[['sex', 'species_id']])
```

```python
df = pd.concat([df, pd.get_dummies(df[['sex', 'species_id']])], axis=1)
df.info()
```

```python
del df['sex']
del df['species_id']
df.info()
```

```python
df
```

#### Let's integrate with missing values handling

```python
filling_values = {"weight": 0, "hindfoot_length": df['hindfoot_length'].mean()}
df = df.fillna(value=filling_values)
df
```
