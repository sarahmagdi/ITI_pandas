```python
import pandas as pd
```

```python
df = pd.read_csv("../datasets/surveys_3.csv")
df.info()
```

```python
df.head()
```

## Equal-Width Discretization

```python
pd.cut(df['weight'], 8)
```

```python
pd.cut(df['weight'], 8, 
       labels=['weight_range1', 'weight_range2', 'weight_range3', 'weight_range4',
               'weight_range5', 'weight_range6', 'weight_range7', 'weight_range8'])
```

```python
pd.get_dummies(
    pd.cut(df['weight'], 8, 
       labels=['weight_range1', 'weight_range2', 'weight_range3', 'weight_range4', 
               'weight_range5', 'weight_range6', 'weight_range7', 'weight_range8'])
)
```

```python
pd.get_dummies(
    pd.cut(df['weight'], 8, 
       labels=['weight_range1', 'weight_range2', 'weight_range3', 'weight_range4', 
               'weight_range5', 'weight_range6', 'weight_range7', 'weight_range8'])
).tail(50)
```

```python
df_weight_discretized_equal_width = pd.concat([df, 
                                               pd.get_dummies(pd.cut(df['weight'], 8, 
                                               labels=['weight_range1', 'weight_range2', 
                                                       'weight_range3', 'weight_range4', 
                                                       'weight_range5', 'weight_range6', 
                                                       'weight_range7', 'weight_range8'])
                                            )], axis=1)
del df_weight_discretized_equal_width['weight']
df_weight_discretized_equal_width
```

```python
df_weight_discretized_equal_width.info()
```
