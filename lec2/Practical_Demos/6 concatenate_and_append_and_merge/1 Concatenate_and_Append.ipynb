```python
import pandas as pd
```

### concat()

```python
df1 = pd.read_json('../datasets/file1.json')
df2 = pd.read_json('../datasets/file1.json')
```

```python
df1
```

```python
df2
```

```python
df = pd.concat([df1, df2])
df
```

```python
df = pd.concat([df1, df2], ignore_index=True)
df
```

```python
df_2 = df1.concat(df2)
# concat can't be called from DataFrame, but append function can do
```

### append()


```python
df_2 = df1.append(df2)
df_2
```

```python
df_2 = df1.append(df2, ignore_index=True)
df_2
```

```python
df_3 = df1.append(df2, ignore_index=True, axis=1)
## append function can't be worked based on columns, but concat function can do
```
